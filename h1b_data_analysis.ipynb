### Analysis the growth of total job titles related to data science job profile over Years.

```python
new_df=df[(df['CASE_STATUS']=='CERTIFIED') | (df['CASE_STATUS']=='DENIED')]
```

```python
new_df.head()
```

```python
dict_cs = {"CERTIFIED" : 1, "DENIED": 0}
dict_fp = {"Y" : 1, "N" : 0}
try:
    
    new_df['CASE_STATUS'] = new_df['CASE_STATUS'].apply(lambda x: dict_cs[x])
    new_df['FULL_TIME_POSITION'] = new_df['FULL_TIME_POSITION'].apply(lambda x: dict_fp[x])
except:
    pass
```

```python
new_df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
categorical_features=['EMPLOYER_NAME','SOC_NAME','JOB_TITLE','WORKSITE']
encoder=LabelEncoder()
new_df[categorical_features] = new_df[categorical_features].apply(encoder.fit_transform)
```

```python
new_df.head()
```

```python
x=new_df[['EMPLOYER_NAME','SOC_NAME','JOB_TITLE','WORKSITE','FULL_TIME_POSITION','PREVAILING_WAGE','YEAR']]
y=new_df['CASE_STATUS']
```

### Modeling

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
import sklearn.metrics as metrics
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.20, random_state=1)
```

```python
lr_clf = LogisticRegression()
lr_clf.fit(X_train, y_train)  
```

```python
y_pred_lr = lr_clf.predict(X_test)

probs = lr_clf.predict_proba(X_test)

print("test", y_test[:10])
print("pred", y_pred_lr[:10])
```

```python
print(metrics.confusion_matrix(y_test,y_pred_lr))
print(metrics.classification_report(y_test, y_pred_lr))
```

```python
metrics.accuracy_score(y_test,y_pred_lr)
```
